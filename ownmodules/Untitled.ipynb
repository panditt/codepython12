```python
import numpy as np

def numpyArray():
    x = np.array([[1, 2, 3], [4, 5, 6]], np.int32)
    y = np.array([[3, 6, 2], [9, 12, 8]], np.int32)
    return x*y
print(numpyArray())
```

```python
def collatz():
    number = input('enter the number whose you want collatz series')
    num = int(number)
    if num % 2 == 0:
        return num // 2
    else:
        return 3 * num + 1

print(collatz())
```
