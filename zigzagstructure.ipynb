```python
import os, random
import time, sys
indent = 0 #for the number of spaces in increasing then decreasing order
indentincreasing = True

try:
    while True: # The main program loop.
        print(' ' * indent, end='')
        print('********')
        time.sleep(0.1) # Pause for 1/10 of a second.

        if indentincreasing == True:
            # Increase the number of spaces:
            indent = indent + 1
            if indent == 20:
                # Change direction:
                indentincreasing = False

        else:
            # Decrease the number of spaces:
            indent = indent - 1
            if indent == 0:
                # Change direction:
                indentincreasing = True
except KeyboardInterrupt:
    sys.exit()
```
